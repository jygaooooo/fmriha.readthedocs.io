# Installation

``fMRI-HA`` is installable via ``pip``:

```python
pip install fmriha
```

Alternatively, you can install the most up-to-date version from GitHub:

```python
git clone https://github.com/SinofWarth/FMRI-HA_TEST.git
cd FMRI-HA_TEST
pip install .
```

Core dependencies are declared in setup.py and requirements.txt, including nilearn>=0.12.0.

If you want to use the executable directly, you can download `fMRI-HA` from xxx.
